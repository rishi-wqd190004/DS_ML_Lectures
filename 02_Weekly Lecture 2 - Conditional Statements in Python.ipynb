## 1. Arithmetic Operators

```python
# Arithmetic Operator
a = 5.0
b = 6
print((float(a))+int(b)) #--> Addition
print(a+b)
print(type(a+b))
print(a-b) #--> Subtraction
print(a*b) #--> Multiplication
print(a/b) #--> Division
print(a//b) #--> Floor Division
```

```python
# Comparison Operators
a = 45
b = 63
 
# a > b is False
print(a > b) # 45 > 63
 
# a < b is True
print(a < b)
 
# a == b is False
print(a == b)
 
# a != b is True
print(a != b) 
 
# a >= b is False
print(a >= b)
 
# a <= b is True
print(a <= b)
```

```python
# Bitwise operators
a = 54
b = 24
c= True
d = False
print(c & d)
 
# Print bitwise AND operation
print(a & b) #and print(a and b)
# print(x & y)
 
# Print bitwise OR operation
print(a | b)
 
# Print bitwise NOT operation
print(~a)
 
# print bitwise XOR operation
print(a ^ b)
 
# print bitwise right shift operation
print(a >> 2)
 
# print bitwise left shift operation
print(a << 2)
```

# For Loop
### For loops are used for sequential traversal.
### For example: traversing a list or string or array etc. 
### In Python, there is “for in” loop which is similar to for each loop in other languages. 

```python
# Syntax:

for iterator_var in sequence:
    statements(s)
```

```python
print(range(0,10))
```

```python
# Example 1
for x in range(0, 10):
    print(x)
print('end')
```

```python
# Example 2 
string = "Capabl"
for i in string:
    print(i)
```

# Nested Loops

### Python programming language allows to use one loop inside another loop. 
### Following section shows few examples to illustrate the concept. 

```python
for iterator_var in sequence:
    for iterator_var in sequence:
        statements(s)
    statements(s)
```

```python
for i in range(1, 5):
    print('i=',i)
    for j in range(0,i):
        print('j=',j, end=' ')
    print('hello')
```

```python
for i in range(1, 5):
    # print(i)
    for j in range(0,i):
        print(j, end=' ')
    print()
```

```python
for i in range(1,5):
    for j in range(i):
        print('*',end=' ')
    print('')
```

# Continue Statement
## The continue statement in Python returns the control to the beginning of the loop.

```python
for letter in 'We are Capabl':
    if letter == 'W' or letter == 'C':
        continue
    print('Current Letter :', letter)
```

# Break Statement
## The break statement in Python brings control out of the loop.

```python
for letter in 'We are Capabl':
 
    # break the loop as soon it sees 'e'
    # or 's'
    if letter == 'W' or letter == 'C':
        break
        
print('Current Letter :', letter)
```

# While Loop in Python
## In python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied. 
## And when the condition becomes false, the line immediately after the loop in the program is executed.

```python
# Syntax:

while expression:
    statement(s)
```

```python
# Python program to illustrate while loop

count = 0
while (count <= 3):
    count = count + 1 # count += 1
    print(count)
    print("We are Capabl")
```

```python
count = 0
while (count < 3):
    count = count + 1
    print('count= ',count)
    print("Capabl")
else:
    print("In Else Block")
```
